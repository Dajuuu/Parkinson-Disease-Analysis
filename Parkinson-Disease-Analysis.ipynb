# 1. Aims, objectives and background

## 1.1 Introduction

Parkinson's disease (PD) is a hypokinetic syndrome characterized by tremors, muscle stiffness, difficulty initiating movement or slowing down of movement. This is because the brain stops producing dopamine. Many factors can cause this disease, but it is not yet established what exactly is the main cause.
The disease itself is not fatal, but symptoms can greatly affect the daily functioning of an affected person, including difficulties in speaking, walking, and interactions with everyday objects.

This is an interesting topic because it is one of the fastest neurological conditions progressing in the world. It does not just affect older people (60+ years). It is estimated that over 10 million people [7] in the world suffer from PD (about 145,000 in the UK [1]). In my project, I would like to look at it and try to answer a few questions.

I tried to find out if this type of analysis had been done before and according to my research no one has yet compared particular countries with each other and even more, no one has done any kind of graphic visualization for a such comparison. All the data gathered is country and region-focused.

## 2.2 What percentage of people suffer from Parkinson's disease in UK countries?

I will start off by making a small analysis on what is the percentage of people suffering from Parkinson's disease in UK countries and the UK as a whole.
This will help to understand how frequent the sickness is appearing in those areas and with a help of the bar chart the distribution of those values can be seen more clearly. The data is taken from the year 2018.

```python
# Load the dataset
parkinsonUK2018 = pd.read_csv("parkinson_uk2018.csv")
parkinsonUK2018.head(7) # check if the data is loaded correctly
```

```python
# Getting the number of occurrences for all countries and their population
# by reading given positions (rows and columns) for the required data
# The given result from the division is then rounded to 4 decimal points

# England
prevalenceEng = parkinsonUK2018.iloc[1,1]
populationEng = parkinsonUK2018.iloc[0,1]
prevalenceEngPerc = round(prevalenceEng/populationEng * 100, 4) 

# Northern Ireland
prevalenceNorthIrl = parkinsonUK2018.iloc[1,2]
populationNorthIrl = parkinsonUK2018.iloc[0,2]
prevalenceNorthIrlPerc = round(prevalenceNorthIrl/populationNorthIrl * 100, 4)

# Scotland
prevalenceScot = parkinsonUK2018.iloc[1,3]
populationScot = parkinsonUK2018.iloc[0,3]
prevalenceScotPerc = round(prevalenceScot/populationScot * 100, 4)

# Wales
prevalenceWales = parkinsonUK2018.iloc[1,4]
populationWales = parkinsonUK2018.iloc[0,4]
prevalenceWalesPerc = round(prevalenceWales/populationWales * 100, 4)

# Whole UK
prevalenceUK = parkinsonUK2018.iloc[1,5]
populationUK = parkinsonUK2018.iloc[0,5]
prevalenceUKPerc = round(prevalenceUK/populationUK * 100, 4)
```

#### Coding the plot to see the percentages

```python
# Saving all of the percentages values to a list 
percentages = [prevalenceEngPerc, prevalenceNorthIrlPerc, prevalenceScotPerc, prevalenceWalesPerc, prevalenceUKPerc]

# Data consists of country name and the calculated prevalence for every UK country
data=[["England", percentages[0]], 
      ["Northern\nIreland", percentages[1]], 
      ["Scotland", percentages[2]],
      ["Wales", percentages[3]], 
      ["United\nKingdom", percentages[4]]]

# Creating dataframe
df=pd.DataFrame(data,columns=["Country","Percentage (%)"])

df.plot(x="Country", y="Percentage (%)", kind="bar", legend=None, figsize=(12,7), color=list("ccccr"))

# Chart information and styling
plt.title("Parkinson's disease in UK countries in 2018 (percentage)", fontsize=16)
plt.xticks(rotation='horizontal')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Percentage", fontsize=14)

# Add the percentage values at the top of every bar 
for index, value in enumerate(percentages):
    plt.text(index, value, str(value) + "%", ha='center', va='bottom', fontweight='bold')
```

Number of population in the UK in 2018 was (as seen in the table above) 66,465,641. By doing small calculations it can be seen that 1 in every 457 people in the UK is afflicted with this disease (!). ## 2.3 How number of cases for different age groups change over the years (with future predictions)?

In the report available on the website Parkinsons.org.uk, predictions on the prevalence for different age groups can be found. By using a line chart I want to see how much of an increase would be across the years and via various age groups. This way it can be estimated, whether the number of people with PD will increase over the years or not, and which age groups will be more affected than others.

CSV file is made out of the report that is available on the website. I also combined the data from the previous dataset.

```python
# Load the dataset
parkinsonUK = pd.read_csv("parkinson_uk2015-2065.csv")
parkinsonUK.head() # check if the data is loaded correctly
```

```python
# Reading the columns in the .csv file and saving them as individual lists
columns = defaultdict(list)

with open("parkinson_uk2015-2065.csv") as f:
    reader = csv.reader(f, delimiter=",")
    next(reader)
    for row in reader:
        for (i,v) in enumerate(row):
            columns[i].append(v)

# numOfPpl - numberOfPeople
# I will use this short version of the variables to be more transparent

# Declaring lists for all values for a given year
numOfPpl2015, numOfPpl2018, numOfPpl2025, numOfPpl2035, numOfPpl2045, numOfPpl2055, numOfPpl2065 = [],[],[],[],[],[],[]

# List of lists (columns)
numOfPpl = [numOfPpl2015, numOfPpl2018, numOfPpl2025, numOfPpl2035, numOfPpl2045, numOfPpl2055, numOfPpl2065]

# Appending the records of first column in the file as age groups strings
ageGroups = columns[0]

# Reading all other columns and saving them to separate variables
for i in range (len(numOfPpl)):
    numOfPpl[i] = columns[i+1]                 # Adding certain columns to a given list
    numOfPpl[i] = list(map(int, numOfPpl[i]))  # Converting list of strings to list of ints 
```

#### Coding the line chart and using the gathered data

```python
# All needed data
data = {"People in 2015": numOfPpl[0], "People in 2018": numOfPpl[1],
        "People in 2025": numOfPpl[2], "People in 2035": numOfPpl[3],
        "People in 2045": numOfPpl[4], "People in 2055": numOfPpl[5],
        "People in 2065": numOfPpl[6], "AgeGroups": ageGroups
       }

# Creating dataframe
df = pd.DataFrame(data, columns=["People in 2015", "People in 2018", "People in 2025","People in 2035", 
                                 "People in 2045", "People in 2055", "People in 2065", "AgeGroups" ])

#Iterate through the length of ageGroup list to display all age groups on the x axis properly  
iterate = []
for i in range(len(ageGroups)):
    iterate.append(i)
    
# Plotting the graph
df.plot(figsize=(14, 12), marker="o")
plt.legend(loc="upper left")
plt.grid(visible=True, linestyle = '--')

# x and y axis information
plt.title("Number of people with Parkinson's disease in different age groups", fontsize=16)
plt.xlabel("Age Groups", fontsize=14)
plt.ylabel("Number of people", fontsize=14)
plt.xticks(iterate, ageGroups) 

plt.show()
```

## 2.4 Does different latitude and longitude affects the occurrences of Parkinson's disease?

From the analysis performed above, the number of patients with PD will be increasing over the years. Previously, the prevalence percentage for different countries for the UK was shown. But will it be this way only for the UK? 
Here is what I will try to analyse. 

I will compare the rates in the UK, Australia and North America and try to determine whether the prevalence rate will be in different places on the Earth. I will check the rates for different years in those countries and by using a bar chart it can be easily determined whether the prevalence percentage will be increasing or not.

However, I was not able to find data from the same years in all of the countries, so I will compare the years that could be found and associated with their information.
I will compare:
- United Kingdom - years <b>2015</b> and <b>2018</b>
- North America - years <b>2010</b> and <b>2020</b>
- Australia - years <b>2011</b> and <b>2014</b>

If data about the population in those years was not provided in the reports used, the population was taken from Google Search Engine. When writing a proper phrase (e.g. population Australia 2011) the exact population for the needed country can be obtained. Data for Google was provided by The World Data Bank [9].

```python
# Load the dataset
parkinsonUK_NA_AU = pd.read_csv("parkinson_uk_na_au.csv")
parkinsonUK_NA_AU.head() # check if the data is loaded correctly
```

```python
# Getting the number of occurrences for all countries and their population
with open('parkinson_uk_na_au.csv') as f:
    reader = csv.reader(f, delimiter=",")
    rows = list(reader)
    
    earlierYear_population = rows[1]  # read the "Populaton - earlier year" row
    earlierYear_population.pop(0)     # getting rid of the "Country" column
    earlierYear_population = list(map(int, earlierYear_population)) # converting list of strings to list of floats
    
    earlierYear_cases = rows[2]       # read the "Number of cases - earlier year" row
    earlierYear_cases.pop(0)          # getting rid of the "Country" column
    earlierYear_cases = list(map(int, earlierYear_cases)) # converting list of strings to list of floats
    
    laterYear_population = rows[3]    # read the "Populaton - later year" row 
    laterYear_population.pop(0)       # getting rid of the "Country" column
    laterYear_population = list(map(int, laterYear_population)) # converting list of strings to list of floats
    
    laterYear_cases = rows[4]         # read the "Number of cases - later year" row
    laterYear_cases.pop(0)            # getting rid of the "Country" column
    laterYear_cases = list(map(int, laterYear_cases)) # converting list of strings to list of floats
    
# lists to collect calculated percentages for all countries 
earlierYear_prevalence = []
laterYear_prevalence = []

# append all calculated values to given lists
for i in range (len(earlierYear_population)):
    earlierYear_prevalencePercentage = round(earlierYear_cases[i]/earlierYear_population[i] * 100, 4)
    earlierYear_prevalence.append(earlierYear_prevalencePercentage)
    
    laterYear_prevalencePercentage = round(laterYear_cases[i]/laterYear_population[i] * 100, 4)
    laterYear_prevalence.append(laterYear_prevalencePercentage)
```

```python
# Creating dataframe
df = pd.DataFrame({
    # Unfortunately, countries with associated years need to be declared explicitly, 
    # as I did not declared those in .csv file, to make the file more readable
    'Country with years': ['United Kingdom\n 2015-2018', 'North America\n 2010-2020', 'Australia\n 2011-2014'],
    'Earlier year': earlierYear_prevalence,   
    'Later year': laterYear_prevalence        
})
  
# Plotting graph
df.plot(x="Country with years", y=["Earlier year", "Later year"], kind="bar", figsize=(12,7), color=list("gr"), zorder=10)
plt.grid(axis='y', zorder=-2.0)
plt.xticks(rotation='horizontal')
plt.title("Prevalence of Parkinson's disease for different countries (percentage)", fontsize=16)  
plt.xlabel("Country", fontsize=14)
plt.ylabel("Percentage", fontsize=14)
```

The previous and the following year have a gap of 10 years in the case of North America, where the biggest difference is present. In UK and Australia, those differences are not so noticeable as the gap between years is not as big. 

Although the years in which the data from those countries is taken are not the same, there is an upward trend in the prevalence rate.

This means there will be an increase in people with Parkinson's disease all over the world and different latitudes and longitudes do not affect the occurrence of the disease.

## 2.5 What kind of symptoms can be seen and which are the most common?

According to the analysed data, the number of people affected by Parkinson's disease will increase over the years (in different parts of the world).

Having this in mind, I will try to find out what symptoms are the most frequent. This section is based on data statistics from Canada.

In general, neurological conditions affect many aspects of life in a bad way. In a study carried out by the Public Health Agency of Canada, the percentages of various symptoms for different, most popular neurological conditions, including Parkinson's disease can be seen.

Using the circle packing chart, it will visualise what symptoms, among the surveyed people with PD, are occurring most often. I would like to analyse what percentage of people suffer from certain symptoms.

```python
# Load the dataset
parkinsonSymptomsCanada = pd.read_csv("parkinson_canada_symptoms2011-2012.csv")
parkinsonSymptomsCanada.head(16) # check if the data is loaded correctly
```

```python
# Collecting the data 
with open('parkinson_canada_symptoms2011-2012.csv') as f:
    csv_reader = csv.reader(f, delimiter=",")
    rows = list(csv_reader)

    neuroConditionsNames = rows[0]  # read the row with names of the neurogical conditions
    neuroConditionsNames.pop(0)     # getting rid of the "Neurological conditions" column
    
    parkinsonSymptoms = rows[13]    # read the row with information about PD
    parkinsonSymptoms.pop(0)        # getting rid of the "Neurological conditions" column
    parkinsonSymptoms = list(map(float, parkinsonSymptoms)) # converting list of strings to list of floats
```

```python
# Creating dataframe
df = pd.DataFrame({'Symptoms':  neuroConditionsNames,
                   'Percentage': parkinsonSymptoms})

# Declaring circles variable to properly show the graph
circles = circlify.circlify(df['Percentage'].tolist(),
                            show_enclosure=False,
                            target_enclosure=circlify.Circle(x=0, y=0, r=1))

# Sorting the percentage values
df = df.sort_values('Percentage')  

fig, ax = plt.subplots(figsize=(12, 12))
ax.set_title('Most common symptoms accompanying PD in Canada in 2011-2012', fontsize=16)
ax.axis('off') # Remove axes


# Find axis boundaries
lim = max(max(abs(circle.x) + circle.r, 
              abs(circle.y) + circle.r)
          for circle in circles)

ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)

# Saving the data to individual variables for easier data manipulation
labels = df['Symptoms']  
percentages = df['Percentage']

# Colours are based on the percentage; the biggest percentage has different colour
colors = ['#eb8a97' if val == df['Percentage'].max() else '#f2b1ec' for val in df['Percentage']]

# Display all of the circles
for per, circle, label, color in zip(percentages, circles, labels, colors):
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha=0.8, linewidth=3, color=color))
    ax.annotate((label + "\n " + str(per) + "%"),   # label - name of a symptom
                (x, y),                             # coordinates
                va='center', ha='center',           # centering
                size=12,                            # size of box
                bbox=dict(facecolor='white', alpha=0.4, edgecolor='white', boxstyle='round', pad=.6)) #box styling
```

As seen above one of the most common symptoms of Parkinson's disease is impaired cognition. People with PD have memory problems. Sometimes they find it difficult to concentrate.

Not so far behind are bladder incontinence and impaired mobility. About the latter, on PD-related websites, many articles put emphasise why appropriate exercises are necessary and important to help slow down the progression of the disease. If you want to more know I refer you to see an article dedicated to this topic, provided by parkinsons.org.uk [10].
